```python
# Exclusion list for non-numeric or unfit columns
exclude_columns = [
    'Student_ID',
    'Date_Range_Week 6',
    'Initial Access Date/Time Assignments',
    'Initial_Learning_Materials',
    'Initial_Module_Info',
    'Initial_Reading_List',
    'Initial_Week_1',                                                                  
    'Initial_Week_2',                                                                  
     'Initial_Week_3',                                                                 
     'Initial_Week_4',                                                                  
     'Initial_Week_5',                                                                   
     'Initial_Week_6', 
     'Initial_Week_7', 
    'Q12 List of Countries',
    'Q10 How do you describe yourself? - Selected Choice',
    'Q16 What is your first degree subject area?',
    'Avg course & student time &logins Student Time in Course'
]

# Process each relevant column
for column in week_6.columns:
    if column in exclude_columns:
        continue

    try:
        # Convert column to numeric
        col_numeric = pd.to_numeric(week_6[column], errors='coerce')
        clean_col = col_numeric.dropna()

        if clean_col.empty:
            continue

        # Calculate bounds
        lower_limit = clean_col.quantile(0.01)
        upper_limit = clean_col.quantile(0.99)

        # Identify original outliers
        original_outliers = week_6[(col_numeric < lower_limit) | (col_numeric > upper_limit)][[column]]

        if not original_outliers.empty:
            print(f"Outliers in column: {column}")
            print(original_outliers)
            print("-" * 30)
            print(f"{column}: values capped to [{lower_limit:.2f}, {upper_limit:.2f}]\n")

        # Apply winsorization
        week_6[column] = col_numeric.clip(lower=lower_limit, upper=upper_limit)

    except Exception as e:
        print(f"⚠️ Skipping column '{column}' due to error: {e}")
```
